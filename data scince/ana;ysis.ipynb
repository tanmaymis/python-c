**actual processing of data set**
drop the columns that are not needed
rename the columns
rename country with very long name
convert all 

```python
cols_to_drop =['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drop,inplace = True)
```

```python
df.head(2)
```

```python
df.rename({
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName': 'Status',
    }, axis=1, inplace=True)
```

```python
df.columns = df.columns.astype(str)   #convert column names to strings 
```

```python
years = list(map(str,range(1980,2014)))     #list of years
```

```python
df[years].sum(axis=1)
```

```python
df['total'] = df[years].sum(axis=1)
```

```python
df.set_index('Country', inplace=True)
```

visualization

```python
df.loc['Japan', years].plot(kind='line', figsize=(15,6), color='red', linewidth=3, marker= 'o')
```

```python
df.loc['India',years].plot(kind ='line')
```

**line,area,bar,barh,pie,scatter,histogram,boxplot**

```python
countries = ['Japan' , 'France']
df.loc[countries,years].T.plot(kind ='line',
                                figsize=(15,6),
                                color =['red','blue'],
                                linewidth=3,
                                marker ='o',
                                markersize=5,
                                )
plt.show()
```

```python
#grouping data by continent
continent_df =df.groupby('Continent').sum()
```

```python
#filter data by continent
filter_europe = df['Continent'] == 'Europe'
eu_df = df[filter_europe].copy()
```

```python
eu_df
```

```python
eu_df.sort_values(by ='total' ,ascending =False , inplace =True)
```

```python
from matplotlib import markers


eu_df.head()[years].T.plot(kind='line',
                            figsize=(15,5),
                            marker='o',
                            markersize=5,
                            title = 'Top countries with imigration in Europe',
                            )
```

```python
eu_df.sort_values(by ='total' ,ascending =False , inplace =True)
eu_df.head()[years].T.plot(kind='line',
                            figsize=(15,5),
                            marker='o',
                            markersize=5,
                            title = 'Top countries with imigration in Europe',
                            subplots = True,
                            layout =(2,3)
                            )
```

```python
df.sort_values(by ='total' ,ascending =False , inplace =True)
df.head(6)[years].T.plot(kind='area',
                            figsize=(20,6),
                            linewidth=2,
                            title = 'Top countries with imigration in world',
                            subplots = True,
                            layout =(2,3)
                            )
plt.show()
```

```python
df.sort_values(by ='total' ,ascending =False , inplace =True)
df.head(6)[years].T.plot(kind='area',
                            figsize=(20,6),
                            linewidth=2,
                            title = 'Top countries with imigration in world',
                            stacked =False
                            )
```
