**actual processing of data set**

```python
cols_to_drop =['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drop,inplace = True)
```

```python
df.head(2)
```

```python
df.rename([])
```

```python
df.rename({
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName': 'Status',
    }, axis=1, inplace=True)
```

```python
df.columns = df.columns.astype(str)   #convert column names to strings 
```

```python
years = list(map(str,range(1980,2014)))     #list of years
```

```python
df[years].sum(axis=1)
```

```python
df['total'] = df[years].sum(axis=1)
```

```python
df.set_index('Country', inplace=True)
```
